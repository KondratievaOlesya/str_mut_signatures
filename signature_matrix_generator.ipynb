```python
import pandas as pd
```

```python
mutation_counts = pd.read_csv("../data/muto_intl/mutation_matrix_all.csv", index_col=0)
mutation_counts
```

```python
# Step 1: Create a mapping from old column names to new simplified mutation types
def extract_mutation_type(col_name):
    parts = col_name.split("_")
    repeat_unit = parts[1]
    change = parts[2]
    unit_len = len(repeat_unit)
    return f"{unit_len}_{change}"

# Apply the function to all columns
new_col_names = mutation_counts.columns.to_series().apply(extract_mutation_type)

# Step 2: Assign new column names and group
mutation_counts.columns = new_col_names

# Step 3: Group by mutation type and sum across columns
grouped_mutation_matrix = mutation_counts.groupby(level=0, axis=1).sum()
```

```python
# Function to extract 'unit' and 'change', ignoring the reference count
def extract_unit_change(col_name):
    parts = col_name.split("_")
    unit = parts[1]
    change = parts[2]
    return f"{unit}_{change}"

# Apply this transformation
unit_change_cols = mutation_counts.columns.to_series().apply(extract_unit_change)

# Rename and group
mutation_counts.columns = unit_change_cols
grouped_by_unit_change = mutation_counts.groupby(level=0, axis=1).sum()

# Show result
grouped_by_unit_change.to_csv("../data/muto_intl/mutation_matrix_ru_change.csv")
grouped_by_unit_change.head()
```

```python
mutation_counts = pd.read_csv("../data/muto_intl/mutation_matrix_all.csv", index_col=0)
mutation_counts
# Function to extract reference count and change only
def extract_ref_change(col_name):
    parts = col_name.split("_")
    ref_count = parts[0]
    change = parts[2]
    return f"{ref_count}_{change}"

# Apply to column names
ref_change_cols = mutation_counts.columns.to_series().apply(extract_ref_change)

# Rename and group
mutation_counts.columns = ref_change_cols
grouped_by_ref_change = mutation_counts.groupby(level=0, axis=1).sum()

# Show result
grouped_by_ref_change.to_csv("../data/muto_intl/mutation_matrix_ref_change.csv")
grouped_by_ref_change.head()
```

```python
mutation_counts = pd.read_csv("../data/muto_intl/mutation_matrix_all.csv", index_col=0)
mutation_counts

# Function to classify unit as AC-rich or not
def is_ac_rich(unit):
    return all(base in {'A', 'T'} for base in unit)

# Function to create the new mutation type
def extract_ac_class(col_name):
    parts = col_name.split("_")
    unit = parts[1]
    change = parts[2]
    unit_len = len(unit)
    ac_label = "ATrich" if is_ac_rich(unit) else "nonAT"
    return f"{unit_len}-{ac_label}_{change}"

# Apply to column names
ac_class_cols = mutation_counts.columns.to_series().apply(extract_ac_class)

# Rename and group
mutation_counts.columns = ac_class_cols
grouped_by_ac_class = mutation_counts.groupby(level=0, axis=1).sum()

# Show result

grouped_by_ac_class.to_csv("../data/muto_intl/mutation_matrix_at_rich_change.csv")
grouped_by_ac_class.head()
```

```python
# Load the mutation matrix
mutation_counts = pd.read_csv("../data/muto_intl/mutation_matrix_all.csv", index_col=0)

# Function to extract RU length
def get_ru_length(col_name):
    unit = col_name.split("_")[1]
    return len(unit)

# Filter out columns with RU length == 1
filtered_columns = [col for col in mutation_counts.columns if get_ru_length(col) > 1]
filtered_mutation_counts = mutation_counts[filtered_columns]
filtered_mutation_counts = filtered_mutation_counts.loc[:, (filtered_mutation_counts != 0).sum(axis=0) > 10]

# Save the filtered matrix
filtered_mutation_counts.to_csv("../data/muto_intl/mutation_matrix_no_ru1.csv")
filtered_mutation_counts.head()
```

```python
from glob import glob
import os
str_matrix_dir = "../data/muto_intl/matricies/"
output_root = "../data/muto_intl/filtered_matricies/"
total_sums = {}

# Loop through each STR matrix file
for matrix_file in glob(f"{str_matrix_dir}/*.csv"):
    matrix_name = os.path.basename(matrix_file).replace(".csv", "")
    
    # Load STR matrix
    X_all = pd.read_csv(matrix_file, index_col=0)
    
    # Calculate total sum of all values
    total_sum = X_all.sum().sum()
    total_sums[matrix_name] = total_sum

    # Filter and save matrix
    filtered_matrix = X_all.loc[:, (X_all != 0).sum(axis=0) > 10]
    filtered_matrix.to_csv(f"{output_root}/filtered_{matrix_name}.csv")

# Print total sums
print("Total sum of all values per matrix:")
for name, value in total_sums.items():
    print(f"{name}: {value}")

# Check if all sums are equal
unique_sums = set(round(v, 6) for v in total_sums.values())
if len(unique_sums) == 1:
    print("\n✅ All original matrices have the same total sum of values.")
else:
    print("\n❌ Matrices differ in their total sum of values.")
```
